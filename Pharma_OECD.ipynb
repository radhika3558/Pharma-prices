In this Notebook, I am filtering OECD pharmaceutical data and creating separate CSVs for each prescription medication to visualize their prices

```python
import pandas as pd
```

```python
df = pd.read_csv('Country_Data_iFHP_Cost_Survey_2017 (1).csv')
df
```

```python
new_df=df[df['Service Type'] == 'Prescription Drugs']
new_df
```

```python
print(new_df.columns)
new_df = new_df.drop(['Year', 'Service Category','Service Type'], axis=1)
```

```python
new_df
```

```python
services = [
    'Xeljanz (Tofacitinib Citrate)',
    'Humira Pen (Adalimumab)',
    'Victoza 3-Pak (Liraglutide) ',
    'Xarelto (Rivaroxaban)',
    'Kalydeco (Ivacaftor)',
    'Enbrel (Etanercept) ',
    'Harvoni (Ledipasvir/Sofosbuvir) '
]

# Create a dictionary to store separate DataFrames for each service
service_dfs = {}

for service in services:
    # Replace parentheses with underscores
    service_key = service.replace('(', '_').replace(')', '_').replace('/', '_')
    service_dfs[service_key] = new_df[new_df['Service'] == service]

# Save each DataFrame to a separate CSV file
for service, df in service_dfs.items():
    file_name = f"{service.replace(' ', '_')}.csv"
    df.to_csv(file_name, index=False)
    print(f"{file_name} saved successfully.")
```
